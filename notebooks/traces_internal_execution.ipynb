```python
import glob, os
import pandas as pd

pd.set_option('display.max_colwidth', None)

path = r'C:\Users\Utente\Desktop\LAND_decoded_calls'
all_files = glob.glob(os.path.join(path, "*.csv"))

internal_df = pd.concat((pd.read_csv(f,  sep=",", engine="python", escapechar='\\')
               for f in all_files))

print(internal_df.shape[0])
print(internal_df.columns)

# df = pd.read_csv(r'C:\Users\Utente\Desktop\LAND_decoded_calls\LAND_decoded_calls_0_0_0.csv', sep=",", engine="python", escapechar='\\')

transactions_df = pd.read_json(r'C:\Users\Utente\Desktop\pm-decentraland\data\transactions\full_land_proxy_transactions.json')

print(transactions_df.shape[0])
print(transactions_df.columns)
```

```python
group_df = internal_df.groupby("TX_HASH")

```

```python
# filter the DFs on same block number ranges

f_internal_df = internal_df[internal_df['BLOCK'].between(14492022, 14494040)]

print(f_internal_df.shape[0])

f_transactions_df = transactions_df[transactions_df['blockNumber'].between(14492022, 14672443)]

print(f_transactions_df.shape[0])
```

```python
from pm4py.objects.log.util import dataframe_utils
from pm4py.objects.conversion.log import converter as log_converter
from pm4py.objects.log.exporter.xes import exporter as xes_exporter

df = f_internal_df.copy()

df = dataframe_utils.convert_timestamp_columns_in_df(
    df)
df = df.sort_values(
    by=["TIMESTAMP", "ORDER_INDEX"])

# remove unnecessary fields
df.drop(["LOAD_ID", "CHAIN_ID", "VALUE", "RAW_ARGUMENTS", "RAW_OUTPUTS", "GAS_USED", "DECODING_STATUS", "STORAGE_ADDRESS"], axis=1, inplace=True)

# drop null values (in case any)
df.dropna(inplace=True)

# create columns: from -> case:concept:name, inputFunctionName -> concept:name, timeStamp -> time:timestamp, from -> org:resource
df["org:resource"] = df["FROM_ADDRESS"]
df["case:concept:name"] = df["TX_HASH"]
df["time:timestamp"] = df["TIMESTAMP"]
df["concept:name"] = df["FUNCTION_NAME"]

# specify that the field identifying the case identifier attribute is the field with name 'case:concept:name'
parameters = {
    log_converter.Variants.TO_EVENT_LOG.value.Parameters.CASE_ID_KEY: 'case:concept:name'}
log = log_converter.apply(df, parameters=parameters,
                          variant=log_converter.Variants.TO_EVENT_LOG)

xes_exporter.apply(
    log, "../data/internal_trace/land_proxy_internal.xes")
```
